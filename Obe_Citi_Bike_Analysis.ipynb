# Bike Repositioning - Using 'for' Loop

```python
#%%timeit
# for a particular bike
from datetime import datetime
bike_trips_df_14529 = pd.DataFrame(bike_trips_df.loc[bike_trips_df['bikeid'] == 14529]).reset_index(drop = True)
bike_trips_df_14529['repos'] = 'no'
bike_trips_df_14529['repostime'] = ''
bike_trips_df_14529['lcolor'] = 'blue'
bike_trips_df_14529['ccolor'] = 'Cesium.Color.BLUE'
for i in range(len(bike_trips_df_14529)):
    if i+1 <= len(bike_trips_df_14529)-1:
        if (bike_trips_df_14529.loc[i,'end station id'] != bike_trips_df_14529.loc[i+1,'start station id']) and \
        (bike_trips_df_14529.loc[i,'repos'] != 'arc') and (bike_trips_df_14529.loc[i+1,'repos'] != 'arc'):
            bike_trips_df_14529.loc[i+1,'repos'] = 'yes'
            bike_trips_df_14529.loc[i+1,'repostime'] = datetime.strptime(bike_trips_df_14529.loc[i+1,'stoptime'],'%Y-%m-%d %H:%M:%S').time()
            bike_trips_df_14529.loc[len(bike_trips_df_14529),:] = bike_trips_df_14529.loc[i+1,:]
            bike_trips_df_14529.loc[len(bike_trips_df_14529)-1,'repos'] = 'arc' 
            bike_trips_df_14529.loc[len(bike_trips_df_14529)-1,'start station id'] = bike_trips_df_14529.loc[i,'end station id']
            bike_trips_df_14529.loc[len(bike_trips_df_14529)-1,'start station name'] = bike_trips_df_14529.loc[i,'end station name']
            bike_trips_df_14529.loc[len(bike_trips_df_14529)-1,'start station latitude'] = bike_trips_df_14529.loc[i,'end station latitude']
            bike_trips_df_14529.loc[len(bike_trips_df_14529)-1,'start station longitude'] = bike_trips_df_14529.loc[i,'end station longitude']
            bike_trips_df_14529.loc[len(bike_trips_df_14529)-1,'lcolor'] = 'red'
            bike_trips_df_14529.loc[len(bike_trips_df_14529)-1,'ccolor'] = 'Cesium.Color.RED'
            bike_trips_df_14529.loc[len(bike_trips_df_14529)-1,'end station id'] = bike_trips_df_14529.loc[i+1,'start station id']
            bike_trips_df_14529.loc[len(bike_trips_df_14529)-1,'end station name'] = bike_trips_df_14529.loc[i+1,'start station name']
            bike_trips_df_14529.loc[len(bike_trips_df_14529)-1,'end station latitude'] = bike_trips_df_14529.loc[i+1,'start station latitude']
            bike_trips_df_14529.loc[len(bike_trips_df_14529)-1,'end station longitude'] = bike_trips_df_14529.loc[i+1,'start station longitude']
            bike_trips_df_14529.loc[len(bike_trips_df_14529)-1,'starttime'] = bike_trips_df_14529.loc[i,'stoptime']
            bike_trips_df_14529.loc[len(bike_trips_df_14529)-1,'stoptime'] = bike_trips_df_14529.loc[i+1,'stoptime']
            #bike_trips_df_14529.loc[len(bike_trips_df_14529)-1,'tripduration'] = bike_trips_df_14529.loc[len(bike_trips_df_14529)-1,'stoptime'].values - bike_trips_df_14529.loc[len(bike_trips_df_14529)-1,'starttime'].values
            
```

```python
# odID is an abbreviation for “origin/ destination identifier”. 
# It is a group identifier. Arc segments which are part of the same origin/destination share the same odID.

# In one cell, we'll create our assignments dataframe.

# Make sure we're starting with an empty dataframe:
assignments_14529 = vrv.initDataframe('assignments')

# Copy over the static values.
# We'll start by copying a single column, to avoid the size mis-match issue:
assignments_14529['objectID'] = bike_trips_df_14529['bikeid']
assignments_14529[['startLat', 'startLon', 'endLat', 'endLon', 'leafletColor', 'cesiumColor']] = bike_trips_df_14529[['start station latitude', 
                                                                          'start station longitude',
                                                                          'end station latitude',
                                                                          'end station longitude','lcolor', 'ccolor']].values

# Copy our new calculated column:
assignments_14529['startTimeSec'] = bike_trips_df_14529['timeAfterStart'].values

# Use the calculated column and tripduration to get the end time (in seconds):
assignments_14529['endTimeSec'] = (bike_trips_df_14529['timeAfterStart'] + bike_trips_df_14529['tripduration']).values

# Fill in the rest of our assignments df with some hard-coded values:
# (we'll probably want to revisit this later)
assignments_14529.loc[:,['modelFile', 'modelScale', 'modelMinPxSize', 'startAltMeters', 'endAltMeters', 
                   'leafletWeight', 'leafletStyle', 'leafletOpacity', 'useArrows',
                   'cesiumWeight', 'cesiumStyle', 'cesiumOpacity']] = \
                  ['veroviz/models/car_blue.gltf', 100, 45, 0, 0, 
                   2, 'solid', 0.8, False, 
                   2, 'solid', 0.7]

# Finally (for now), let's generate a unique odID value for each row.
# This will make sense only if we assume that each row corresponds to a specific
# O/D pair.  Conversely, if we have turn-by-turn arcs, we'll need to group
# multiple rows into the same O/D pair.  We'll tackle that case if/when 
# we encounter it.
assignments_14529.loc[:,'odID'] = list(range(0, len(assignments_14529)))
```

```python
#%%timeit

# for a particular bike
from datetime import datetime

# original assignment dataframe
bike_trips_df_14529D = pd.DataFrame(bike_trips_df.loc[bike_trips_df['bikeid'] == 14529]).reset_index(drop = True)
bike_trips_df_14529D['repos'] = 'no'
bike_trips_df_14529D['repostime'] = ''
bike_trips_df_14529D['lcolor'] = 'blue'
bike_trips_df_14529D['ccolor'] = 'Cesium.Color.BLUE'

#create dataframes for comparing
bike_trips_df_14529A = pd.DataFrame(bike_trips_df.loc[bike_trips_df['bikeid'] == 14529]).reset_index(drop = True)
# rename columns
bike_trips_df_14529A.columns = ['tripdurationA', 'starttimeA', 'stoptimeA', 'start station idA',
       'start station nameA', 'start station latitudeA',
       'start station longitudeA', 'end station idA', 'end station nameA',
       'end station latitudeA', 'end station longitudeA', 'bikeidA', 'usertypeA',
       'birth yearA', 'genderA']

bike_trips_df_14529B = pd.DataFrame(bike_trips_df_14529A.loc[1:,:]).reset_index(drop = True)
# rename columns
bike_trips_df_14529B.columns = ['tripdurationB', 'starttimeB', 'stoptimeB', 'start station idB',
       'start station nameB', 'start station latitudeB',
       'start station longitudeB', 'end station idB', 'end station nameB',
       'end station latitudeB', 'end station longitudeB', 'bikeidB', 'usertypeB',
       'birth yearB', 'genderB']

# drop last row in A since it will not have another row for comparison
bike_trips_df_14529A.drop(bike_trips_df_14529A.tail(1).index,inplace=True)

# combine dataframes to compare for repositioning assignment
bike_trips_df_14529C = pd.concat([bike_trips_df_14529A, bike_trips_df_14529B], axis=1, sort=False).reset_index(drop = True)

# add columns for repositioning assignement and create calculated fields
bike_trips_df_14529C['tripduration'] = bike_trips_df_14529C['tripdurationA']
bike_trips_df_14529C['starttime'] = bike_trips_df_14529C['stoptimeA']
bike_trips_df_14529C['stoptime'] = bike_trips_df_14529C['starttimeB']
bike_trips_df_14529C['start station id'] = bike_trips_df_14529C['end station idA']
bike_trips_df_14529C['start station name'] = bike_trips_df_14529C['end station nameA']
bike_trips_df_14529C['start station latitude'] = bike_trips_df_14529C['end station latitudeA'] 
bike_trips_df_14529C['start station longitude'] = bike_trips_df_14529C['end station longitudeA']
bike_trips_df_14529C['end station id'] = bike_trips_df_14529C['start station idB']
bike_trips_df_14529C['end station name'] = bike_trips_df_14529C['start station nameB']
bike_trips_df_14529C['end station latitude'] = bike_trips_df_14529C['start station latitudeB']
bike_trips_df_14529C['end station longitude'] = bike_trips_df_14529C['start station longitudeB']
bike_trips_df_14529C['bikeid'] = bike_trips_df_14529C['bikeidA']
bike_trips_df_14529C['usertype'] = 'Staff'
bike_trips_df_14529C['birth year'] = 'Staff'
bike_trips_df_14529C['gender'] = 'Staff'

# new columns
bike_trips_df_14529C['repos'] = 'arc'
bike_trips_df_14529C['repostime'] = ''
bike_trips_df_14529C['lcolor'] = 'red'
bike_trips_df_14529C['ccolor'] = 'Cesium.Color.RED'

# only keep arcs with repositioning occurring
bike_trips_df_14529E = pd.DataFrame(bike_trips_df_14529C.loc[bike_trips_df_14529C['end station idA'] != bike_trips_df_14529C['start station idB']]).reset_index(drop = True)
bike_trips_df_14529E = bike_trips_df_14529E[['tripduration', 'starttime', 'stoptime', 'start station id',
       'start station name', 'start station latitude',
       'start station longitude', 'end station id', 'end station name',
       'end station latitude', 'end station longitude', 'bikeid', 'usertype',
       'birth year', 'gender', 'repos', 'repostime',
       'lcolor', 'ccolor']]

#join original dataframe (D) with repositioning dataframe (E)
bike_trips_df_14529F = pd.concat([bike_trips_df_14529D, bike_trips_df_14529E],ignore_index=True).reset_index(drop = True)
```
